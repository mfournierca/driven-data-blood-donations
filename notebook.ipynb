## First Model: Logistic Regression

Our success metric as defined by the contest is the log loss, which is the cost function minimized by the logistic regression classifier. 

We use the logistic classifier in an sklearn pipeline and a grid search to find the best hyperparameters. 

### Learning Curves

Converges around 0.5, need a better model. 

Recall that higher scores are better by convention so the log loss is inverted: http://scikit-learn.org/stable/modules/model_evaluation.html

```python
from src import model
log_learning = model.get_learning_curve(model.logistic(), x, y)
ggplot(log_learning, aes(x="sizes")) + geom_line(aes(y="train_score"), colour="red") + geom_line(aes(y="cv_score"), colour="blue")
```

### Polynomial Features

```python
m = model.logistic(max_poly_degree=3)
m.get_params()["estimator__polynomial_features__degree"]
```

```python
log_learning = model.get_learning_curve(m, x, y)
ggplot(log_learning, aes(x="sizes")) + geom_line(aes(y="train_score"), colour="red") + geom_line(aes(y="cv_score"), colour="blue")
```
