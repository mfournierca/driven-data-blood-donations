```python
from ggplot import *
from src import data, model
x, y = data.load_all()
x.columns
```

```python
ggplot(aes(x="Months since Last Donation", y="Number of Donations"), data=x) + geom_point()
```

```python
log_learning = model.get_learning_curve(model.logistic, x, y, stepsize=25)
log_learning.head()
```

```python
ggplot(log_learning, aes(x="n_train")) + geom_line(aes(y="train_log_loss"), colour="red") + geom_line(aes(y="test_log_loss"), colour="blue")
```

```python
gnb_learning = model.get_learning_curve(model.gnb, x, y, stepsize=25)
gnb_learning.head()
```

```python
ggplot(gnb_learning, aes(x="n_train")) + geom_line(aes(y="train_log_loss"), colour="red") + geom_line(aes(y="test_log_loss"), colour="blue")
```
