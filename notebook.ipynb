## First Model: Logistic Regression

Our success metric as defined by the contest is the log loss, which is the cost function minimized by the logistic regression classifier. 

We use the logistic classifier in an sklearn pipeline and a grid search to find the best hyperparameters. 

### Learning Curves

High variability, need more data for this model. 

```python
log_learning = model.get_learning_curve(model.logistic, x, y, stepsize=25)
ggplot(log_learning, aes(x="n_train")) + geom_line(aes(y="train_log_loss"), colour="red") + geom_line(aes(y="test_log_loss"), colour="blue")
```
